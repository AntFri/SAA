```python

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Create a sample DataFrame

# calcular el mapa de corelacion
corr_matrix = df.corr()

# Creacion del Mapa
fig, ax = plt.subplots(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', ax=ax)

```
